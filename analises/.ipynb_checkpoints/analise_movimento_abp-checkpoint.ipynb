```python
import numpy as np
import pandas as pd
import altair as alt
import sys
import json
alt.data_transformers.disable_max_rows()

sys.path.insert(0, '../agents')
sys.path.insert(0, '../environments')
sys.path.insert(0, '..')

import ps_env_abp_target_find as env_class
import ps_agent as agent_class
import projective_simulation_iteration as ps_model
```

```python
model_path = '../results/models/2024-05-04_20-22-12'
model = ps_model.ProjectiveSimulation.load(model_path)
model
```

```python
steps = 100000

# Se reseta o ambiente no início de cada episódio


# Inicializa variáveis para armazenar posição e estados
position = []
var_position = []
states = []
state_counter = []

# Inicia passos de aprendizado do episódio
for i in range(10):
    model.reset_environment()
    model.reset_agent()
    model.env.reset_rng()

    # Setando posição fixa do agente e do alvo
    target_position = np.array([0, 0])
    agent_position = np.array([50.0, 50.0])

    model.set_environment_atribute('r', agent_position)
    model.set_environment_atribute('target_position', target_position)
    state_ += 1
    for step in range(steps):
        position.append(model.get_environment_atribute('r'))
        #var_position.append(model.get_environment_atribute('dr_dt'))
        _ = model.get_environment_atribute('state_observation')()

        done = model.run_learning_step()

        if len(state_counter) == 0:
            state_ = 0

        elif states[-1][0] != _[0]:
            state_ += 1
            
        state_counter.append(state_)
        states.append(_)
        
        if done: # Se encontrou recompensa, finaliza episódio
            print('Done: {} steps'.format(step))
            break
```

```python
# Gerando dataframe de estados e posições
db = pd.concat([
    pd.DataFrame(position).rename(columns={0: 'x', 1: 'y'}), 
    #pd.DataFrame(var_position).rename(columns={0: 'dx',1:'dy'}),
    pd.DataFrame(states).rename(columns={0: 'state', 1: 'timer'}), 
    pd.DataFrame(state_counter).rename(columns={0: 'state_counter'})
], axis = 1)

db['x_target'] = target_position[0]
db['y_target'] = target_position[1]

db['dx'] = db.groupby(['state_counter'])['x'].diff().fillna(0)
db['dy'] = db.groupby(['state_counter'])['y'].diff().fillna(0)

db['dx'] = db['dx'].apply(lambda x: min(abs(x), 100-abs(x)))
db['dy'] = db['dy'].apply(lambda x: min(abs(x), 100-abs(x)))

db
```

# Estados ativos: state == 1

```python
db_state = db.groupby(['state_counter','state'])[['dx','dy']].sum().apply(
    lambda x: np.sqrt(x['dx']**2 + x['dy']**2), axis=1
    ).reset_index().rename(columns={0: 'distance'}).merge(
        db[['state_counter','timer']].groupby('state_counter')['timer'].max().reset_index())

db_state['v'] = db_state['distance']/db_state['timer']
db_state
```

## Animação

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Create a figure and axis
plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['figure.dpi'] = 300
plt.ioff()
fig, ax = plt.subplots()

# Initialize the scatter plot
scat = ax.scatter(
    db['x'], 
    db['y'], 
    c=db['state'], 
    cmap='viridis', 
    s=50
)

ax.axis([0, 100, 0, 100])

frame_tail = 50
# Define the update function for the animation
def update(frame):
    if frame > frame_tail:
        scat.set_offsets(np.column_stack((
            db['x'][frame - frame_tail:frame], 
            db['y'][frame - frame_tail:frame]
        )))
        scat.set_array(db['state'][frame - frame_tail:frame])
    else:
        scat.set_offsets(np.column_stack((db['x'][:frame], db['y'][:frame])))
        scat.set_array(db['state'][:frame])
    return scat,

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=500, interval=30, blit=True)

ani.save("teste.gif", dpi=300, writer=animation.PillowWriter(fps=25))
#from IPython.display import HTML
#HTML(ani.to_jshtml())
```
